```python
# libraries
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.integrate as spi
```

1. Write a program to plot a line graph representing the temperature readings over a week: 20,22,19,23,21,24,20

```python
# lineplot
data = [20,22,19,23,21,24,20] # temperature readings
fig, ax = plt.subplots(figsize=(10,5))
sns.lineplot(data=data, ax=ax, color="g")
ax.set_xlabel("Day of Week")
ax.set_ylabel("Temperature")
ax.set_title("Temperature Over Week")
ax.legend(["Temperature"])
```

2. Write a program to generate the arithmetic sequence starting at 5 with a common difference of 3, for 8 terms

```python
numbers = [] # list to hold the values
start = 5 # stat point
for number in range(8):
  number = start + number * 3
  numbers.append(number) # add the values to the list 'numbers'

print(numbers)
```

3. Write a program to calculate the volume under the surface z = $x^{2}$ + $y^2$ over the square region 0  x,y  1

```python
# define the function z = x^2 + y^2
def integrand(x, y):
  return x**2 + y**2

# set the integrand
x_lower = 0
x_upper = 1
y_lower = lambda x:0
y_upper = lambda x:1

# perform the double integration
volume, error = spi.dblquad(integrand, x_lower, x_upper, y_lower, y_upper)

# print the result
print(f"Volume under the surface is: {round(volume, 4)} with error: {error}")
```

4. Explain the difference between compiled and interpreted programming languages, and classify Python in this context

* The main difference between compiled and interpreted programming languages is in how codes are run: with compiled languages, the code is converted into machine code ahead of time which speeds up the process, while with interpreted languages the code is run step by step once the program runs, making the process slower but more flexible.  
* Python is seen as an interpreted language, even though it does turn source code into bytecode before running the code. Contrary to machine code which computer processors can run directly, this bytecode is a platform-independent version and it is run on the Python Virtual Machine (PVM). This mixed system gives Python ease of debugging and the ability to run on different platforms, along with some extra speed because of the compiled bytecode.
